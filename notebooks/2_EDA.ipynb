```python
single_opt = ['EdLevel','DevType','Country','AISelect']
multi_opt = ['Employment','CodingActivities','LanguageHaveWorkedWith','DatabaseHaveWorkedWith',
             'PlatformHaveWorkedWith','WebframeHaveWorkedWith','MiscTechHaveWorkedWith']
number = ['YearsCode','YearsCodePro','WorkExp']
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load & subset
df = pd.read_csv("../data/raw/survey_results_public.csv")
cols_of_interest = single_opt + multi_opt + number + ['CompTotal']
df = df[cols_of_interest].copy()

# 2. Clean & convert
# 2a. Compensation
df['CompTotal'] = pd.to_numeric(df['CompTotal'], errors='coerce')
# 2b. Years fields
years_map = {'Less than 1 year': 0.5, 'More than 50 years': 50}
for col in number:
    df[col] = df[col].replace(years_map)
    df[col] = pd.to_numeric(df[col], errors='coerce')

# 3. Preliminary summary
print("=== Data shape ===")
print(df.shape)
print("\n=== Missing values ===")
print(df.isna().mean().sort_values(ascending=False).head(10))

# 4. Single‐choice analysis
for col in single_opt:
    print(f"\n## {col} ##")
    vc = df[col].fillna("Missing").value_counts(normalize=True).mul(100).round(1)
    print("Top categories (%):")
    print(vc.head(5))
    med_sal = (
        df.groupby(col)['CompTotal']
          .median()
          .dropna()
          .sort_values(ascending=False)
    )
    print("\nTop 5 median salaries:")
    print(med_sal.head(5))
    plt.figure(figsize=(6,4))
    sns.barplot(x=med_sal.head(10).values, y=med_sal.head(10).index)
    plt.title(f"Median Salary by {col}")
    plt.xlabel("Median Compensation")
    plt.tight_layout()
    plt.show()

# 5. Numeric variables summary & correlation
print("\n## Numeric summary ##")
print(df[number + ['CompTotal']].describe().T)

corr = df[number + ['CompTotal']].corr()['CompTotal'].sort_values(ascending=False)
print("\nCorrelations with CompTotal:")
print(corr)

plt.figure(figsize=(4,4))
sns.heatmap(df[number + ['CompTotal']].corr(), annot=True, fmt=".2f", cmap="vlag")
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()

# 6. Multi‐choice analysis
for col in multi_opt:
    print(f"\n## {col} ##")
    exploded = (
        df[col]
        .dropna()
        .str.split(';')
        .explode()
        .str.strip()
    )
    counts = exploded.value_counts().head(10)
    print("Top 10 options by count:")
    print(counts)
    med_sal = (
        df.assign(**{col: df[col].str.split(';')})
          .explode(col)[['CompTotal', col]]
          .dropna(subset=[col])
          .assign(**{col: lambda d: d[col].str.strip()})
          .groupby(col)['CompTotal']
          .median()
          .sort_values(ascending=False)
    )
    print("\nTop 5 median salaries:")
    print(med_sal.head(5))
    plt.figure(figsize=(6,4))
    sns.barplot(x=med_sal.head(10).values, y=med_sal.head(10).index)
    plt.title(f"Median Salary by {col} Option")
    plt.tight_layout()
    plt.show()

# 7. In‐depth example: Highest‐paying programming languages
lang_med = (
    df.assign(Language=df['LanguageHaveWorkedWith'].str.split(';'))
      .explode('Language')
      .dropna(subset=['Language'])
      .assign(Language=lambda d: d['Language'].str.strip())
      .groupby('Language')['CompTotal']
      .median()
      .sort_values(ascending=False)
)
print("\n=== Top 10 Highest‐Paying Languages (median) ===")
print(lang_med.head(10))

plt.figure(figsize=(6,4))
sns.barplot(x=lang_med.head(10).values, y=lang_med.head(10).index)
plt.title("Top 10 Median Salaries by Programming Language")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

df = pd.read_csv("../data/raw/survey_results_public.csv")
# Set the style for plots
sns.set_style('whitegrid')

# Analyze the target variable: ConvertedCompYearly
plt.figure(figsize=(10, 6))
sns.histplot(df['ConvertedCompYearly'].dropna(), bins=100, kde=True)
plt.title('Distribution of Annual Salary (USD)')
plt.xlabel('Salary (USD)')
plt.ylabel('Frequency')
plt.show()

# We'll likely need to handle the long tail of high salaries. A log transformation can help.
df['LogSalary'] = np.log1p(df['ConvertedCompYearly'])

plt.figure(figsize=(10, 6))
sns.histplot(df['LogSalary'].dropna(), bins=100, kde=True)
plt.title('Distribution of Log-Transformed Annual Salary (USD)')
plt.xlabel('Log Salary')
plt.ylabel('Frequency')
plt.show()


# Explore relationships with key categorical features
def plot_feature_salary(feature, top_n=15):
    plt.figure(figsize=(12, 8))
    # Taking the top N most frequent categories to keep the plot readable
    top_categories = df[feature].value_counts().nlargest(top_n).index
    sns.boxplot(data=df[df[feature].isin(top_categories)], x=feature, y='ConvertedCompYearly')
    plt.title(f'Salary Distribution by {feature}')
    plt.xlabel(feature)
    plt.ylabel('Annual Salary (USD)')
    plt.xticks(rotation=45, ha='right')
    plt.show()

plot_feature_salary('Country')
plot_feature_salary('EdLevel')
plot_feature_salary('DevType')


# Correlation matrix for numerical features
plt.figure(figsize=(10, 8))
numerical_cols = df.select_dtypes(include=np.number).columns
correlation_matrix = df[numerical_cols].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
