```python
import numpy as np
import pandas as pd
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import (
    OrdinalEncoder, StandardScaler
)

class YearsCodeConverter(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.special_map = {
            "Less than 1 year": 0.5,
            "More than 50 years": 51
        }

    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        X_copy = X.copy()
        for col in X_copy.columns:
            X_copy[col] = X_copy[col].replace(self.special_map)
            X_copy[col] = pd.to_numeric(X_copy[col], errors='coerce')
        return X_copy

    def get_feature_names_out(self, input_features=None):
        return input_features

# --- Updated Preprocessor Builder ---
def build_preprocessor():
    """
    Builds a full, modular scikit-learn pipeline that now includes
    the custom conversion for numeric columns.
    """
    numeric_features = ['YearsCode', 'YearsCodePro', 'WorkExp']
    categorical_features = [
        'EdLevel', 'DevType', 'Country', 'AISelect', 'Employment', 
        'CodingActivities', 'LanguageHaveWorkedWith', 'DatabaseHaveWorkedWith',
        'PlatformHaveWorkedWith', 'WebframeHaveWorkedWith', 'MiscTechHaveWorkedWith'
    ]

    numeric_pipeline = Pipeline(steps=[
        ('converter', YearsCodeConverter()),
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ])

    categorical_pipeline = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='constant', fill_value='Missing')),
        ('ordinal', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
    ])
    
    preprocessor = ColumnTransformer(
        transformers=[
            ('numeric', numeric_pipeline, numeric_features),
            ('categorical', categorical_pipeline, categorical_features)
        ],
        remainder='drop'
    )
    return preprocessor
```

```python
import numpy as np
import pandas as pd
import json
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import (
    OrdinalEncoder, StandardScaler, OneHotEncoder
)
import joblib

# --- Configuration ---
OHE_CARDINALITY_THRESHOLD = 25

class YearsCodeConverter(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.special_map = {
            "Less than 1 year": 0.5,
            "More than 50 years": 51
        }

    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        X_copy = X.copy()
        for col in X_copy.columns:
            if X_copy[col].dtype == 'object':
                X_copy[col] = X_copy[col].replace(self.special_map)
            X_copy[col] = pd.to_numeric(X_copy[col], errors='coerce')
        return X_copy

    def get_feature_names_out(self, input_features=None):
        return input_features

def build_preprocessor(numeric_features, ohe_features, ordinal_features):
    numeric_pipeline = Pipeline(steps=[
        ('converter', YearsCodeConverter()),
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ])

    ohe_pipeline = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='constant', fill_value='Missing')),
        ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
    ])

    ordinal_pipeline = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='constant', fill_value='Missing')),
        ('ordinal', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1, dtype=np.int64))
    ])
    
    preprocessor = ColumnTransformer(
        transformers=[
            ('numeric', numeric_pipeline, numeric_features),
            ('onehot', ohe_pipeline, ohe_features),
            ('ordinal', ordinal_pipeline, ordinal_features)
        ],
        remainder='drop',
        n_jobs=-1
    )
    return preprocessor

print("Loading raw data...")
df_raw = pd.read_csv("../data/raw/survey_results_public.csv", low_memory=False)

# --- Define Feature Sets ---
all_features = [
    'YearsCode', 'YearsCodePro', 'WorkExp', 'EdLevel', 'DevType', 
    'Country', 'AISelect', 'Employment', 'CodingActivities', 
    'LanguageHaveWorkedWith', 'DatabaseHaveWorkedWith', 'PlatformHaveWorkedWith',
    'WebframeHaveWorkedWith', 'MiscTechHaveWorkedWith',
    'OpSysProfessional use' 
]
TARGET = 'ConvertedCompYearly'
numeric_features = ['YearsCode', 'YearsCodePro', 'WorkExp']

# --- Dynamic Feature Categorization ---
categorical_features = [col for col in all_features if col not in numeric_features]
ohe_features = []
ordinal_features = []

for col in categorical_features:
    if df_raw[col].nunique(dropna=False) <= OHE_CARDINALITY_THRESHOLD:
        ohe_features.append(col)
    else:
        ordinal_features.append(col)

print(f"Features to be One-Hot Encoded (<= {OHE_CARDINALITY_THRESHOLD} unique values): {ohe_features}")
print(f"Features to be Ordinal Encoded (> {OHE_CARDINALITY_THRESHOLD} unique values): {ordinal_features}")

# --- Data Cleaning and Preparation ---
df = df_raw[all_features + [TARGET]].copy()
df.dropna(subset=[TARGET], inplace=True)

print("Capping outliers...")
features_to_cap = numeric_features + [TARGET]
for col in features_to_cap:
    if col in ['YearsCode', 'YearsCodePro']:
        special_map = {"Less than 1 year": 0.5, "More than 50 years": 51}
        if df[col].dtype == 'object':
            df[col] = df[col].replace(special_map)
        df[col] = pd.to_numeric(df[col], errors='coerce')

    p25, p75 = df[col].dropna().quantile([0.25, 0.75])
    iqr = p75 - p25
    upper_limit = p75 + 1.5 * iqr
    lower_limit = p25 - 1.5 * iqr
    df[col] = np.where(
        df[col] > upper_limit, upper_limit,
        np.where(df[col] < lower_limit, lower_limit, df[col])
    )

X = df[all_features]
y = df[TARGET]

# --- Preprocessing and Saving ---
preprocessor = build_preprocessor(numeric_features, ohe_features, ordinal_features)
print("Fitting preprocessor on the entire dataset...")
preprocessor.fit(X)

print("Transforming data...")
processed_data_np = preprocessor.transform(X)
try:
    feature_names = preprocessor.get_feature_names_out()
except Exception as e:
    print(f"Could not get feature names from preprocessor, generating them manually. Error: {e}")
    num_features = preprocessor.named_transformers_['numeric'].get_feature_names_out()
    ohe_features_out = preprocessor.named_transformers_['onehot'].get_feature_names_out()
    ord_features_out = preprocessor.named_transformers_['ordinal'].get_feature_names_out()
    feature_names = list(num_features) + list(ohe_features_out) + list(ord_features_out)

X_processed_df = pd.DataFrame(
    processed_data_np, 
    columns=feature_names, 
    index=X.index
)

print("Combining processed features and target variable...")
y_df = y.to_frame()
final_df = pd.concat([X_processed_df, y_df], axis=1)

print("Saving the combined dataset and the fitted preprocessor...")
joblib.dump(preprocessor, 'preprocessor.pkl')
final_df.to_csv("../data/processed/final_dataset.csv")

# --- Generate Schema JSON ---
print("Generating input schema...")

schema = {}

# Numeric Features
for col in numeric_features:
    schema[col] = {
        "type": "numeric",
        "selection": "single",
        "values": None
    }

# Categorical Features
for col in ohe_features + ordinal_features:
    unique_vals = df[col].dropna().unique().tolist()

    if df[col].dropna().astype(str).str.contains(';').any():
        selection_type = "multi"
        all_vals = (
            df[col].dropna()
            .astype(str)
            .str.split(';')
            .explode()
            .str.strip()
            .unique()
            .tolist()
        )
    else:
        selection_type = "single"
        all_vals = [str(val).strip() for val in unique_vals]

    schema[col] = {
        "type": "categorical",
        "selection": selection_type,
        "values": sorted(all_vals)
    }

with open("../data/processed/cat_domains.json", "w") as f:
    json.dump(schema, f, indent=4)

print("Input schema saved to: cat_domains.json")
print("\nProcess complete.")
print(f"Preprocessor saved to: preprocessor.pkl")
print(f"Combined dataset saved to: final_dataset.csv (Shape: {final_df.shape})")
```
