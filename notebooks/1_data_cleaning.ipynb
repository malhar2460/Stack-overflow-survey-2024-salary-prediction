```python
import numpy as np
import pandas as pd
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import (
    OrdinalEncoder, StandardScaler
)

def build_preprocessor():
    """
    Builds a full, modular scikit-learn pipeline for preprocessing the 
    Stack Overflow survey data based on specified column types.
    
    This version uses OrdinalEncoder for all categorical features.
    """
    
    # --- Define Column Groups ---
    numeric_features = ['YearsCode', 'YearsCodePro', 'WorkExp']
    
    # Combine all categorical features into a single list for Ordinal Encoding
    all_categorical_features = [
        'EdLevel', 'DevType', 'Country', 'AISelect', # Single-option
        'Employment', 'CodingActivities', 'LanguageHaveWorkedWith', # Multi-option
        'DatabaseHaveWorkedWith', 'PlatformHaveWorkedWith', 
        'WebframeHaveWorkedWith', 'MiscTechHaveWorkedWith'
    ]

    # --- Define Individual Pipelines for each Data Type ---

    # Pipeline for numerical data: impute missing values with the median, then scale.
    numeric_pipeline = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ])

    # Pipeline for ALL categorical data: impute missing as 'Missing', then apply OrdinalEncoder.
    # This will treat each unique string (including semicolon-separated ones) as a distinct category.
    categorical_pipeline = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='constant', fill_value='Missing')),
        # OrdinalEncoder converts each category into a single integer.
        # handle_unknown is crucial for preventing errors if new categories
        # appear in the data during prediction.
        ('ordinal', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
    ])
    
    # --- Combine All Pipelines into a Single Preprocessor ---
    # The ColumnTransformer now applies the two main pipelines to their respective column groups.
    preprocessor = ColumnTransformer(
        transformers=[
            ('numeric', numeric_pipeline, numeric_features),
            ('categorical', categorical_pipeline, all_categorical_features)
        ],
        remainder='drop' # Drop any columns not specified in our lists
    )
    
    return preprocessor
```

```python
preprocessor = build_preprocessor()
preprocessor.fit(df)
```
