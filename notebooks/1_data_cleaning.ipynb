```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import FunctionTransformer
import pandas as pd

dumper = FunctionTransformer(
    func=lambda s: s.str.get_dummies(sep=";"),
    validate=False
)

col_trans = (
    ColumnTransformer(
        transformers=[
            ("empl_onehot", dumper, "Employment")
        ],
        remainder="passthrough"
    )
    .set_output(transform="pandas")
)

df_encoded = col_trans.fit_transform(df)

print(df_encoded.dtypes)
print(df_encoded.head())
```
